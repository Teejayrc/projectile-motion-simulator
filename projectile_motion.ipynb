```python
import math
import matplotlib.pyplot as plt
```

```python
# Inputs
velocity = float(input("Enter initial velocity (m/s): "))
angle = float(input("Enter angle of projection (degrees): "))
g = 9.81  # Acceleration due to gravity (m/s^2)

# Convert angle to radians
angle_rad = math.radians(angle)

# Calculate time of flight, maximum height, and range
time_of_flight = (2 * velocity * math.sin(angle_rad)) / g
max_height = (velocity**2 * math.sin(angle_rad)**2) / (2 * g)
range_of_projectile = (velocity**2 * math.sin(2 * angle_rad)) / g

# Display results
print(f"Time of Flight: {time_of_flight:.2f} seconds")
print(f"Maximum Height: {max_height:.2f} meters")
print(f"Range: {range_of_projectile:.2f} meters")

# Generate trajectory
t = [i * 0.01 for i in range(int(time_of_flight / 0.01) + 1)]
x = [velocity * math.cos(angle_rad) * i for i in t]
y = [(velocity * math.sin(angle_rad) * i - 0.5 * g * i**2) for i in t]

# Plot the trajectory
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Trajectory")
plt.title("Projectile Motion")
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Distance (m)")
plt.legend()
plt.grid()
plt.show()
```
